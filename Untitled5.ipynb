```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("darkgrid")
```

```python
control_tiempos = pd.read_excel("/content/drive/MyDrive/CC/Gestion de tiempos/Tapa Corona/Datos.xlsx")
datos = control_tiempos[control_tiempos["TIEMPO PARADA (MIN)"] <= 70]
datos = datos[datos["INCLUYE"] == "SI"]
datos.head()
```

```python
print(type(datos["TIEMPO PARADA (MIN)"]))
```

```python
# Gráfica de distribución por Línea
plt.figure(figsize=(15, 6))
#g = sns.FacetGrid(datos, col="MES", sharex = False, palette= "pastel",col_wrap = 4)
#g.map_dataframe(sns.histplot, x="TIEMPO PARADA (MIN)")
#g.map_dataframe(sns.displot, x="TIEMPO PARADA (MIN)", hue="LÍNEA", kind="kde")
sns.displot(data = datos, x="TIEMPO PARADA (MIN)", hue="LÍNEA",kde = True, col = "MES", col_wrap = 4)
plt.show()
```

```python
# Comportamiento línea COMBI por mes
# Datos
datos_combi = datos[datos["LÍNEA"] == "Línea Combi"]
plt.figure(figsize=(15, 6))
sns.lineplot(data = datos_combi, x="FECHA", y="TIEMPO PARADA (MIN)")
plt.title("Comportamiento Línea Combi")
# Mostrar la gráfica
plt.show()
```

```python
# Comportamiento línea COMBI PLUS por mes
# Datos
datos_combi_plus = datos[datos["LÍNEA"] == "Línea Combi Plus"]
plt.figure(figsize=(15, 6))
sns.lineplot(data = datos_combi_plus, x="FECHA", y="TIEMPO PARADA (MIN)")
plt.title("Comportamiento Línea Combi Plus")
# Mostrar la gráfica
plt.show()
```

```python
#Gráfico de calor de línea por producto
datos_TLP = datos[["TIEMPO PARADA (MIN)", "LÍNEA", "PRODUCTO"]]
datos_pivot = datos_TLP.pivot_table(index="PRODUCTO", columns="LÍNEA", values="TIEMPO PARADA (MIN)", aggfunc='sum')

plt.figure(figsize = (10,6))
sns.heatmap(datos_pivot,annot=True,fmt='.0f',cmap="YlGnBu")
plt.title("Correlación de producto vs máquina")
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Crear un DataFrame a partir de los datos proporcionados
data = {
    'Pliegue': [1, 1, 1, 2, 2, 2],
    'Y': [0.35, 7.37, -5.84, 4.29, 4.36, 0.06],
    'X': [-0.05, 0.79, -2.24, 0.83, 0.2, -0.63]
}
df = pd.DataFrame(data)

# Separar los datos en dos pliegues
pliegue1 = df[df['Pliegue'] == 1]
pliegue2 = df[df['Pliegue'] == 2]

# Entrenar y evaluar el modelo para cada pliegue
# Pliegue 1 como prueba, Pliegue 2 como entrenamiento
modelo1 = LinearRegression()
modelo1.fit(pliegue2[['X']], pliegue2['Y'])
predicciones1 = modelo1.predict(pliegue1[['X']])
mse1 = mean_squared_error(pliegue1['Y'], predicciones1)

# Pliegue 2 como prueba, Pliegue 1 como entrenamiento
modelo2 = LinearRegression()
modelo2.fit(pliegue1[['X']], pliegue1['Y'])
predicciones2 = modelo2.predict(pliegue2[['X']])
mse2 = mean_squared_error(pliegue2['Y'], predicciones2)

# Calcular el MSE promedio
mse_promedio = (mse1 + mse2) / 2
mse_promedio

```

```python
import pandas as pd

# Tus registros
registros = ["valores1;valores2", "valores2;valores3"]

# Creamos un conjunto de todos los valores únicos
valores_unicos = set()
for registro in registros:
    valores_unicos.update(registro.split(';'))

# Creamos un diccionario para almacenar los valores y sus presencias
data = {}
for valor in valores_unicos:
    data[valor] = []

# Iteramos sobre los registros y marcamos las presencias
for registro in registros:
    valores_registro = registro.split(';')
    for valor in valores_unicos:
        if valor in valores_registro:
            data[valor].append(1)
        else:
            data[valor].append(0)

# Creamos el DataFrame
df = pd.DataFrame(data)

print(df)
```
